```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
import librosa
import librosa.display
```

```python
duration=2
frequency=5
c_sampling_rate=1000
```

```python
cont_time=np.linspace(0,duration,duration*c_sampling_rate)
cont_x=np.sin(2*np.pi*frequency*cont_time)
plt.figure(figsize=(15,4))
plt.plot(cont_time,cont_x)
```

```python
sampling_rate=50
discrete_time=np.linspace(0,duration,duration*sampling_rate)
sampled_x=np.sin(2*np.pi*frequency*discrete_time)
plt.figure(figsize=(15,4))
plt.stem(discrete_time,sampled_x)
```

```python
def quantization(x,bit):
    width=2/(2**bit)
    y=np.round(x/width)*width
    return y
sq_x=quantization(sampled_x,4)
plt.figure(figsize=(15,4))
plt.step(discrete_time,sq_x)
```

```python
sampling_rate=50
duration=3
frequency2=1
discrete_time=np.linspace(0,duration,duration*sampling_rate)
sampled_x=np.sin(2*np.pi*frequency*discrete_time)+np.cos(2*np.pi*frequency2*discrete_time)
plt.figure(figsize=(15,4))
plt.stem(discrete_time,sampled_x)
```

```python
fft=np.abs(np.fft.fft(sampled_x))
freq=np.fft.fftfreq(len(sampled_x),1/sampling_rate)
plt.figure(figsize=(15,4))
plt.stem(freq,fft)
print(fft)
print(len(fft))
```
